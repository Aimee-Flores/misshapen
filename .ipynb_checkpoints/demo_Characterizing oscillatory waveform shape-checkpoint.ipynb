In our paper (link TBD) we argue that characteristics of the neural oscillation waveform carries information about the underlying neural activity that cannot be obtained with traditional Fourier-based methods.

This notebook provides a demonstration of some of the functions in this library by characterizing the sharpness of some simulated oscillations

Dependencies:
* [tools](https://github.com/srcole/tools)

**Standard Imports**

```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

**Load and visualize data**

```python
lfps = np.load('./lfpall.npy')
N = len(lfps)
S = len(lfps[0])
flo = (13,30)
dt = .001
t = np.arange(0,S*dt,dt)
Smax = 1000
```

```python
plt.figure(figsize=(10,20))
for n in range(N):
    plt.subplot(N,1,n+1)
    plt.plot(t[:Smax],lfps[n][:Smax],'k')
    plt.tight_layout()
    plt.ylabel('V')
plt.xlabel('Time (s)')
```

# Extrema sharpness

```python
# Quantify shape metric as a function of gaussian std
from misshapen.shape import findpt, ex_sharp

# Shape parameters
boundaryS = 100
widthS = 5

pk_sharp = np.zeros(N,dtype=object)
tr_sharp = np.zeros(N,dtype=object)
pk_t = np.zeros(N,dtype=object)
tr_t = np.zeros(N,dtype=object)
for n in range(N):
    pk_t[n], tr_t[n] = findpt(lfps[n], flo, Fs = 1/dt, boundary = boundaryS)
    pk_sharp[n] = ex_sharp(lfps[n], pk_t[n], widthS, Fs=1/dt, fosc = flo)
    tr_sharp[n] = ex_sharp(lfps[n], tr_t[n], widthS, Fs=1/dt, fosc = flo)
```

```python
# Visualize extrema sharpness distributions for 2 simulations
sim1 = 4
sim2 = 9
bins=np.arange(-2,2,.2)

plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.hist(np.log10(pk_sharp[sim1]),bins,color='k',label='peaks',alpha=0.5)
plt.hist(np.log10(tr_sharp[sim1]),bins,color='r',label='troughs',alpha=0.5)
plt.ylabel('Count')
plt.title('Asymmetric oscillations')
plt.legend(loc='best')

plt.subplot(2,1,2)
plt.hist(np.log10(pk_sharp[sim2]),bins,color='k',label='peaks',alpha=0.5)
plt.hist(np.log10(tr_sharp[sim2]),bins,color='r',label='troughs',alpha=0.5)
plt.xlabel('Log Extrema sharpness')
plt.ylabel('Count')
plt.title('Symmetric oscillations')
```

# Extrema sharpness ratio

```python
# Quantify ESR
from misshapen.shape import esr
esrmethod = 'aggregate'

esrlfp = np.zeros(N,dtype=object)
for n in range(N):
    esrlfp[n] = np.log10(esr(lfps[n],pk_t[n],tr_t[n],widthS,esrmethod=esrmethod))
```

```python
plt.figure(figsize=(6,6))
plt.plot(esrlfp,'k.')
plt.ylabel('Log ESR')
plt.xticks([0,N-1],['Asymmetric','Symmetric'])
plt.xlim((0-.5,N-.5))
plt.ylim((0,1))
```
